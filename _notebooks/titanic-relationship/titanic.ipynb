# Titanic Relationship between people
This notebook is only to evaluate the relationship between peoples based on their surname, cabin and ticket.

```python
import pandas as pd
from IPython.display import display 

df=pd.read_csv('train.csv')
df.head()
```

From the dataset we can see the name of the passengers and their cabin, let's see what we can dig?

```python
#Find the cabin with the most people
display(df['Cabin'].value_counts().head(5).reset_index()) #top 5 most populated cabins

most_populated_cabin=df['Cabin'].value_counts().head(1).index[0] #top 1
print('top 1 is',most_populated_cabin)
df[df['Cabin']==most_populated_cabin]
```

```python
#Is it always the first name the surname?
print('Names without comma',df[df['Name'].str.find(',')==-1].shape[0])
#No name without comma 

df['Surname']=df['Name'].str.split(',').str[0]
df.head(5)


```

```python
#Let's check the rank of surnames?
df['Surname'].value_counts().head(5)
```

Who are the Andersson?

```python
df[df['Surname']=='Andersson'].sort_values(by=['Ticket'])
```
